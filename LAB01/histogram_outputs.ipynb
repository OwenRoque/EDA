```python
import matplotlib.pyplot as plt
```

```python
file_names = [
    "distances_dim_10.txt",
    "distances_dim_50.txt",
    "distances_dim_100.txt",
    "distances_dim_500.txt",
    "distances_dim_1000.txt",
    "distances_dim_2000.txt",
    "distances_dim_5000.txt",
]
bin_sizes = [10, 50, 50, 100, 100, 200, 200]
```

```python

# Iterate all files & bin sizes
for file_name, bins in zip(file_names, bin_sizes):
    # Read all distances from the file
    with open(file_name, "r") as file:
        distances = [float(line.strip()) for line in file]

    # Create a histogram
    plt.hist(distances, bins=bins, range=(0, max(distances)), color='skyblue')
    plt.title("Histograma - " + file_name)
    plt.xlabel("Distancia")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()
```
